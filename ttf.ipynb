```python
import PIL
from PIL import ImageFont, Image, ImageDraw

# 生成font
ttf_path = 'Aa剑豪体.ttf'
text_size = 255 # text_size 是字号
font = ImageFont.truetype(ttf_path, text_size)

x,y =font.getsize('三')
print(x,'--',y)
```

```python
if y<256:
    y=256
if x<256:
    x=256
cavv = Image.new('RGB',tuple([x,y]),(255,255,255))

ddd =ImageDraw.Draw(cavv)
ddd.text((0,0),'三',font=font,fill='#000000')
cavv
image=cavv.resize((256,256),Image.ANTIALIAS)
image.save('aaa.png')
image
```

```python
from fontTools.ttLib import TTFont

def get_char_list_from_ttf(font_file):
    ' 给定font_file,获取它的中文字符 '
    f_obj = TTFont(font_file)
    m_dict = f_obj.getBestCmap()

    unicode_list = []
    for key, _ in m_dict.items():
        unicode_list.append(key)

    char_list = [chr(ch_unicode) for ch_unicode in unicode_list]
    return char_list

font_file = 'Aa剑豪体.ttf'
chars = get_char_list_from_ttf(font_file)
print(chars)
```
